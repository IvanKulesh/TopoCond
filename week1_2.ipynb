# Let's try to impliment 1D wire

$$
H = -\frac{\hbar^2}{2m} \partial_x^2
$$

$$
\partial_x^2 \psi(x) = \frac{1}{\epsilon^2} \big( \psi(x+\epsilon) + \psi(x-\epsilon) - 2 * \psi(x) \big)
$$

From this expression it's clear that the matrix representation of the momentum operator is proportional to:

$$
\begin{bmatrix}
    \vdots & \ddots & \vdots & \vdots & \vdots & \vdots\\
    1 & -2 & 1  & 0 & \dots & \dots \\
    0 & 1 & -2 & 1  & 0 & \dots \\
    \dots  & 0 & 1 & -2 & 1 & 0\\
    \vdots & \vdots & \vdots & \vdots & \ddots & \vdots
\end{bmatrix}
$$

Hence we have:
$$
H = -\frac{\hbar^2}{2m} \sum_{i=0}^{N}
\Big(
\left| n+1 \right\rangle  \left\langle n \right|
+ \left| n \right\rangle  \left\langle n+1 \right|
- 2\left| n \right\rangle  \left\langle n \right| 
\Big)
$$

The hopping amplitude is $-t$ and on-site energy is just $2t$
$$
t = \frac{\hbar^2}{2m}
$$

# Su-Schrieffer-Heeger (SSH) model

![](figures/polyacetylene.png)

In this model we consider only hoppings between neighbouring sites.

First we will rewrite the Hamiltonian separating sublattice degree of freedom:

$$
H = \sum_{n=1}^{N} \big ( t_1 c_{n,a}^{\dagger} c_{n,b} + t_2 c_{n,b}^{\dagger} c_{n+1,a} + h.c. \big ),
$$


where N is half the number of atoms.

We can make a vector 
$$
\mathbf{c}_n^{\dagger} =  \big ( c_{n,a}^{\dagger}, c_{n,b}^{\dagger} \big )
$$

So the Hamiltonian will have the form

$$
H =  \sum_{m,n=1}^{N} \mathbf{c}_m^{\dagger} H_{m,n} \mathbf{c}_n
$$

It's straightforward to verify that we have for "onsight" energy

$$
\mathbf{c}_n^{\dagger} H_{n,n} \mathbf{c}_n = t_1 c_{n,a}^{\dagger} c_{n,b} + h.c = \big ( c_{n,a}^{\dagger}, c_{n,b}^{\dagger} \big )
\begin{pmatrix}
0 & t_1\\
t_1 & 0
\end{pmatrix}
\begin{pmatrix}
 c_{n,a}\\
 c_{n,b}
\end{pmatrix} \quad \implies \quad
H_{n,n} = t_1 \sigma_x
$$

and for hopping:

$$
\mathbf{c}_n^{\dagger} H_{n,n+1} \mathbf{c}_{n+1} = t_2 c_{n,b}^{\dagger} c_{n+1,a} = \big ( c_{n,a}^{\dagger}, c_{n,b}^{\dagger} \big )
\begin{pmatrix}
0 & 0\\
t_2 & 0
\end{pmatrix}
\begin{pmatrix}
 c_{n+1,a}\\
 c_{n+1,b}
\end{pmatrix} \quad \implies \quad
H_{n,n+1} = t_2 \frac{\sigma_x - i \sigma_y}{2}
$$

$$
\mathbf{c}_{n+1}^{\dagger} H_{n+1,n} \mathbf{c}_{n} = t_2 c_{n+1,a}^{\dagger} c_{n,b} = \big ( c_{n+1,a}^{\dagger}, c_{n+1,b}^{\dagger} \big )
\begin{pmatrix}
0 & t_2\\
0 & 0
\end{pmatrix}
\begin{pmatrix}
 c_{n,a}\\
 c_{n,b}
\end{pmatrix} \quad \implies \quad
H_{n+1,n} = t_2 \frac{\sigma_x + i \sigma_y}{2}
$$

The Hamiltonian in real space has representation:

$$
\hat{H}_{SSH} = \sum_{n} t_1 \left| n \right\rangle  \left\langle n \right| \otimes \sigma_x
+ \sum_{n} \Big[
t_2  \left| n \right\rangle  \left\langle n + 1 \right| \otimes \frac{\sigma_x - i \sigma_y}{2}
+ 
t_2  \left| n+1 \right\rangle  \left\langle n  \right| \otimes \frac{\sigma_x + i \sigma_y}{2}
\Big]
$$

Comapre this to the Kitaev chain:

$$
\hat{H}_{Kitaev} = \sum_{n}(-\mu) \left| n \right\rangle  \left\langle n \right| \otimes \sigma_z
- \sum_{n} \Big[
\left| n \right\rangle  \left\langle n + 1 \right| \otimes \left ( t \sigma_z + i \Delta \sigma_y \right)
+ h.c.
\Big]
$$


We introduce momentum basis states (note the different than in lectures sign!):
$$
\left| k \right\rangle = \frac{1}{\sqrt{N}}\sum_{n}e^{ikn}\left| n \right\rangle
$$

$$
\hat{H}(k)_{SSH} = t_1 \otimes \sigma_x + t_2 \cos(k) \otimes \sigma_x + t_2 \sin(k) \otimes \sigma_y
$$

for Kitaev chain we had:
$$
\hat{H}(k)_{Kitaev} = -\mu\otimes \sigma_z -2 t \cos(k) \otimes \sigma_z - 2 \Delta \sin(k) \otimes \sigma_y
$$

# About symmetry:

The encountered here symmetry is the Chiral symmetry:

$$
\sigma_z \hat{H}(k)_{SSH} \sigma_z^{\dagger} = - \hat{H}(k)_{SSH}
$$
This means that every state has a partner with same momentum and negative energy.

Compare this to
$$
\sigma_x \mathcal{K} \hat{H}(k)_{Kitaev} \mathcal{K}^{\dagger} \sigma_x^{\dagger} = - \hat{H}(-k)_{Kitaev}
$$

# About $\sigma_z$

Imagine we had hoppings between sites with same sublattice (next nearest neighbours):

$$
t_3 c_{n,a}^{\dagger} c_{n+1,a} + t_4 c_{n,b}^{\dagger} c_{n+1,b} = 
\mathbf{c}_n^{\dagger} H_{n,n+1} \mathbf{c}_{n+1} = \big ( c_{n,a}^{\dagger}, c_{n,b}^{\dagger} \big )
\begin{pmatrix}
t_3 & 0\\
0 & t_4
\end{pmatrix}
\begin{pmatrix}
 c_{n+1,a}\\
 c_{n+1,b}
\end{pmatrix} \quad \implies \quad
H_{n,n+1} = g_3 \sigma_0 + g_4 \sigma_z
$$

This will brake the Chiral symmetry, because
$$
\sigma_z \Big( g_3 \sigma_0 + g_4 \sigma_z \Big) \sigma_z^{\dagger} \neq - \Big( g_3 \sigma_0 + g_4 \sigma_z \Big) 
$$

# Kwant simulation:

```python
import kwant
%matplotlib notebook
import matplotlib.pyplot as pyplot
import tinyarray
import scipy.sparse.linalg as sla
import numpy as np
```

```python
sigma_0 = np.array([[1, 0], [0, 1]])
sigma_x = np.array([[0, 1], [1, 0]])
sigma_y = np.array([[0, -1j], [1j, 0]])
sigma_z = np.array([[1, 0], [0, -1]])
```

```python
def hopx(site1, site2, t_1, t_2, g_3, g_4):
    return t_2 * (sigma_x - 1j * sigma_y) / 2.0 + g_3 * sigma_0 + g_4 * sigma_z

def onsite(site, t_1, t_2, g_3, g_4):
    return t_1 * sigma_x
```

```python
a=1
L = 50
t_2 = 1.0
g_3 = 0.00
g_4 = 0.00
```

```python
syst = kwant.Builder()

lat = kwant.lattice.chain(a)

syst[( lat(x) for x in range(L) )] = onsite
syst[lat.neighbors()] = hopx

syst = syst.finalized()
```

```python
t_1_vals_0 = np.linspace(0.5, 2.8, 153)
energies_0 = []

for t_1 in t_1_vals_0:
    ham_mat = syst.hamiltonian_submatrix(args=[t_1, t_2, g_3, g_4], sparse=True)
    ev = sla.eigs(ham_mat, k=16, sigma=0, which='LM', return_eigenvectors=False, maxiter=15000)
    energies_0.append(np.sort(ev))
```

```python
pyplot.figure()
pyplot.plot(t_1_vals_0, energies_0, 'o', markersize='1')
pyplot.xlabel("t1")
pyplot.ylabel("energy")
pyplot.show()
```
